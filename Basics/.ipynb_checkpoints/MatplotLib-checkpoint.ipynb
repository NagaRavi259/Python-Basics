# MatplotLib
- MatplotLibis a plottinglibrary for the python programinglanguage and its numarical mathematics extention Numpy. It provides an object-oeriented API for embedding plots into applications using general purpose GUI toolkit linke Tkinter, wxPython, Qt or GTK+. etc
### Some major pros of MatPlotLib
    - generally esy to get started for simple plots
    - Support for custom labels and texts
    - Great control of every element in a figure
    - High-quality output in many formats
    - Very custmizable in general

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
##Writing this line will show the graph without using the plt.show()
```

```python
x=np.arange(0,10)# created a 1D numpy array witha rangeof 0 to 10
y=np.arange(11,21)# created a 1D numpy array witha rangeof 11 to 21
```

```python
a=np.arange(40,50)# created a 1D numpy array witha rangeof 40 to 50
b=np.arange(50,60)# created a 1D numpy array witha rangeof 50 to 60
```

## Scatter graph
- Scatter() is used to scatter the points in the graph using the values givenin X and Y direction
- c='' takes a string argument and represents the points inside the graph as we specified in the statement 

```python
plt.scatter(x,y,c='b') #c='g' thsi specifies the color of the points
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
plt.title('This is atest graph')
plt.savefig('sctter.jpg')
```

## Plots
- plots graph will show the allpoints into a stright line
    - simply as steight line
- in plots graph we dont  need to pass the c='g' for color we can simply pass third element with a string with a code of color
- -- to get a dashed line insted of stright line
- *- to represent the each point as start in line graph

```python
y=x**3
plt.plot(x,y,'mo-') # in plots graph we dont  need to pass the c='g' for color we can simply pass third element with a string with a code of color
                    # -- to get a dashed line insted of stright line
                    # *- to represent the each point as start in line graph
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
plt.title('This is atest graph')
```

### Subplots
- subplot() 
    - Will create a multiple plots in a single sketch with sized each one  
    - we have pass 3 values inside the sketch 
        - No of Rows
        - No of Columns
        - Possition of the grph

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*-')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'yo-')
```

```python
x=np.arange(1,11)
y=3*x+5
plt.title('Sine wave graph')
plt.plot(x,y)
```

#### To crate  sine wave on a matplotlib

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title('Sine wave Graph')
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
plt.plot(x,y)
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.tan(x)     ## Tan graph 
plt.title('Sine wave Graph')
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
plt.plot(x,y)
```

```python
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
y_tan=np.tan(x)
plt.subplot(3,1,1)
plt.plot(x,y_sin)
plt.subplot(3,1,2)
plt.plot(x,y_cos)
plt.subplot(3,1,3)
plt.plot(x,y_tan)
```

## Bar Plot
- 

```python
x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,17]

plt.bar(x,y, align = 'center')
plt.bar(x2,y2, color='g')
plt.title('Bar graph')
plt.ylabel('Y-axis')
plt.xlabel('X-Axis')
```

## Histogram
- Histogram is similar to the bar graph but it shows densityof the multiple data in the list
- 

```python
a=np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a)
plt.title('Histogram graph')
plt.ylabel('Y-axis')
plt.xlabel('X-Axis')
```

## Box plot
- Box plots helps to get the percentile of the data


```python
data=[np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=True);
```

## Pie Chart
- Pie chart is the representation of category is the data

```python
#Data to plot
labels= 'Python','C++','Ruby','Java' # Labels of the Pie
Sizes=[255,90,285,210] #cumulative total size of the pie
colors= ['gold','yellowgreen','lightcoral','lightskyblue'] # colors of the each tile Pie
explode=(0.1,0,0,0) #  it will define how much fr that tile should move from the pie
#Plotting
plt.pie(Sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)# autopact represents the how many floating points should represent in the pie tile 
plt.axis('equal')
plt.show()
```
